**Number of Rides per Hour**

```python
df=pd.read_csv('/content/uber_trip_analysis.csv')

import seaborn as sns
import matplotlib.pyplot as plt

# 1. Trips per Hour: Identifies peak rush hours
plt.figure(figsize=(12, 6))
sns.countplot(x='Pickup_Time', data=df, palette='viridis')
plt.title('Number of Rides per Hour')
plt.show()



```

**Distribution of Trip Distances**

```python
# 5. Mileage Distribution: Most rides are short-distance
plt.figure(figsize=(10, 6))
sns.histplot(df['Distance(km)'], bins=30, kde=True, color='teal')
plt.title('Distribution of Trip Distances')
plt.xlabel('Distance(km)')
plt.show()

```

**Circular Heatmap (Day vs Hour)**

```python
# Group data to create a pivot table
pivot_table = df.pivot_table(index='Day', columns='Pickup_Time', values='Date', aggfunc='count')

plt.figure(figsize=(15, 8))
sns.heatmap(pivot_table, cmap='YlGnBu', annot=False)
plt.title('Ride Intensity: Day of Week vs Hour of Day')
plt.show()
```

**Violin Plots (Location vs Distance)**

```python
plt.figure(figsize=(12, 8))
sns.violinplot(x='Distance(km)', y='Pickup_Location', data=df, inner="quart", palette="muted")
plt.title('Mileage Distribution by Trip Location')
plt.show()
```

**Pair Plot (Variable Inter-relationships)**

```python
sns.pairplot(df[['Distance(km)', 'Duration(Min', 'Pickup_Time', 'Pickup_Location']], hue='Pickup_Location', diag_kind='kde')
plt.show()
```
